# ðŸ“Š Calculadora de Markup y Descuentos

## ðŸ”¢ CÃ¡lculo de Markup: (100/descuento)-1

Este notebook permite calcular el markup necesario para aplicar descuentos en productos.

```python
# ===================================
# 1ï¸âƒ£ CONFIGURACIÃ“N DE DESCUENTO
# ===================================
valor_equivalente = 75   # Agregar el valor equivalente del descuento aplicado (%)

# CÃ¡lculo del markup necesario
markup = (100 / valor_equivalente) - 1

# Mostrar resultados
print("=" * 60)
print("        ðŸŽ¯ RESULTADO DEL CÃLCULO DE MARKUP")
print("=" * 60)
print(f"\nðŸ“Œ Valor equivalente:       {valor_equivalente}%")
print(f"ðŸ“ˆ Markup para el Admin:    {markup:.4f} ({markup * 100:.2f}%)")
print(f"ðŸ’° Markup aplicado:         {100 - valor_equivalente}%")
print("\n" + "=" * 60)
```

---

## ðŸ’³ CÃ¡lculo Aplicado al Producto (Millas/Puntos)

AquÃ­ calculamos el valor real del producto con y sin markup aplicado.

```python
# ===================================
# 2ï¸âƒ£ CONFIGURACIÃ“N DEL PRODUCTO
# ===================================
requiredPoints =  200569     # Puntos/Millas requeridos
pointValue = 0.01             # Valor de cada punto en USD

# ===================================
# ðŸ“Š CÃLCULOS
# ===================================
# Valor con markup (precio inflado)
valor_usd_con_markup = requiredPoints * pointValue

# Valor sin markup (precio real despuÃ©s del descuento)
valor_markup_aplicado_USD = valor_usd_con_markup * ((100 - valor_equivalente) / 100)

# Diferencia (markup aplicado)
valor_usd_sin_markup = valor_usd_con_markup - valor_markup_aplicado_USD

# ===================================
# ðŸ“„ REPORTE DE RESULTADOS
# ===================================
print("\n" + "=" * 60)
print("        ðŸ’° ANÃLISIS DE PRECIOS")
print("=" * 60)
print(f"\nðŸ”¸ Puntos requeridos:          {requiredPoints:,}")
print(f"ðŸ”¸ Valor por punto:            USD ${pointValue}")
print(f"ðŸ”¸ Markup aplicado:            {100 - valor_equivalente}%")
print("\n" + "-" * 60)
print(f"\nðŸ“ˆ Precio Con markup:-------------->USD ${valor_usd_con_markup:}")
print(f"ðŸ’µ Valor markup:------------------->USD ${valor_markup_aplicado_USD:}")
print(f"ðŸ“ˆ Precio sin Markup aplicado:----->USD ${valor_usd_sin_markup:}")

print("\n" + "=" * 60)
```

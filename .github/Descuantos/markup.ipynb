# ğŸ“Š Calculadora de Markup y Descuentos

## ğŸ”¢ CÃ¡lculo de Markup: (100/descuento)-1

Este notebook permite calcular el markup necesario para aplicar descuentos en productos.

```python
# ===================================
# 1ï¸âƒ£ CONFIGURACIÃ“N DE DESCUENTO
# ===================================
descuento_aplicado = 75  # Agregar el valor equivalente del descuento aplicado (%)

# CÃ¡lculo del markup necesario
markup = (100 / descuento_aplicado) - 1

# Mostrar resultados
print("=" * 60)
print("        ğŸ¯ RESULTADO DEL CÃLCULO DE MARKUP")
print("=" * 60)
print(f"\nğŸ“Œ Valor equivalente:       {descuento_aplicado}%")
print(f"ğŸ“ˆ Markup para el Admin:    {markup:.4f} ({markup * 100:.2f}%)")
print(f"ğŸ’° Descuento real aplicado: {100 - descuento_aplicado}%")
print("\n" + "=" * 60)
```

---

## ğŸ’³ CÃ¡lculo Aplicado al Producto (Millas/Puntos)

AquÃ­ calculamos el valor real del producto con y sin markup aplicado.

```python
# ===================================
# 2ï¸âƒ£ CONFIGURACIÃ“N DEL PRODUCTO
# ===================================
requiredPoints = 200569      # Puntos/Millas requeridos
pointValue = 0.01            # Valor de cada punto en USD

# ===================================
# ğŸ“Š CÃLCULOS
# ===================================
# Valor con markup (precio inflado)
valor_usd_con_markup = requiredPoints * pointValue

# Valor sin markup (precio real despuÃ©s del descuento)
valor_usd_sin_markup = valor_usd_con_markup * ((100 - descuento_aplicado) / 100)

# Diferencia (markup aplicado)
markup_valor = valor_usd_con_markup - valor_usd_sin_markup

# ===================================
# ğŸ“„ REPORTE DE RESULTADOS
# ===================================
print("\n" + "=" * 60)
print("        ğŸ’° ANÃLISIS DE PRECIOS")
print("=" * 60)
print(f"\nğŸ”¸ Puntos requeridos:        {requiredPoints:,}")
print(f"ğŸ”¸ Valor por punto:          USD ${pointValue}")
print(f"ğŸ”¸ Descuento aplicado:       {100 - descuento_aplicado}%")
print("\n" + "-" * 60)
print(f"\nğŸ“ˆ Precio CON markup:        USD ${valor_usd_con_markup:,.2f}")
print(f"ğŸ’µ Precio SIN markup:        USD ${valor_usd_sin_markup:,.2f}")
print(f"ğŸ Descuento (ahorro):       USD ${markup_valor:,.2f}")
print("\n" + "=" * 60)
```
